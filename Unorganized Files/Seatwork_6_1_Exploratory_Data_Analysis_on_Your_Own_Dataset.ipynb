```python
# created a dataframe to check the table for prepared students

column = 'test preparation course'
value = ['completed']

checker = df[column].isin(value) # .isin() is also a function for checking if a column
                                   # has a certain value

prepared = df[checker]

prepared
```

```python
# number of prepared students
print(len(prepared), 'students are prepared')
```

```python
# created a dataframe to check the table for unprepared students

column = 'test preparation course'
value = ['none']

checker = df[column].isin(value) # .isin() is also a function for checking if a column
                                   # has a certain value

unprepared = df[checker]

unprepared
```

```python
# number of prepared students
print(len(unprepared), 'students are not prepared')
```

```python
# created a dataframe to check the number of students who passed math

column = 'math score'
# we assume that 70 is the passing score
min_value = 70
max_value = 100

checker = df[column].between(min_value, max_value) # .between() function for checking between two certain integers

passed_math = df[checker]

passed_math
```

```python
len(passed_math)
```

```python
# create dataframe to see how many who prepared and passed math

column = 'math score'
# we assume that 70 is the passing score
min_value = 70
max_value = 100

checker = prepared[column].between(min_value, max_value) # .between() function for checking between two certain integers

p_passed_math = prepared[checker]

p_passed_math
```

```python
len(p_passed_math)
```

```python
# create dataframe to see how many who did not prepared and passed math

column = 'math score'
# we assume that 70 is the passing score
min_value = 70
max_value = 100

checker = unprepared[column].between(min_value, max_value) # .between() function for checking between two certain integers

u_passed_math = unprepared[checker]

u_passed_math
```
