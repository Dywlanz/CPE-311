```python
m_q1 = df['math score'].quantile(0.25) # first quartile
m_q3 = df['math score'].quantile(0.75) # third quartile
m_iqr = m_q3-m_q1 # IQR
m_min = m_q1-(1.5*m_iqr) # minimum
m_max = m_q3+(1.5*m_iqr) # maximum
print('Math score IQR:', m_iqr)
print('Minimum math score:', m_min)
print('Maximum math score:', m_max)
```
